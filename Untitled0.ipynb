```python
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
search_result = lk.search_lightcurve('Kepler-10', author='Kepler')
search_result
```

```python
lc = search_result.download()
lc.plot()
```

```python
# Clean, normalize and fold the light curve
folded_lc = lc.remove_nans().normalize().fold(period=0.837495)
folded_lc.plot()
```

```python
import matplotlib.pyplot as plt

folded_lc.plot()
plt.xlim(-0.1, 0.1)
plt.show()

```
